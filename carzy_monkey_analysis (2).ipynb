## RFM Analysis

```python
r_cust=returning_customers.copy()
```

```python
r_cust.rename(columns= {'created_at':'order_date'} ,inplace=True)
```

```python
r_cust.head()
```

## Recency
Calculating Recency
Here we are calculating recency for customers who had made a purchase from our website .

```python
df.size_id.unique()
```

```python
df_recency=r_cust.groupby(by='user_id',as_index=False)['order_date'].max()

df_recency.columns = ['user_id', 'LastPurchaseDate']
```

```python

recent_date = df_recency['LastPurchaseDate'].max()

df_recency['Recency'] = df_recency['LastPurchaseDate'].apply(
    lambda x: (recent_date - x).days)
```

```python
df_recency.head()
```

## Frequency
Calculating Frequency
We are here calculating the frequency of frequent transactions of the customer in ordering/buying some product from the company.

```python
frequency_df = r_cust.groupby(
    by=['user_id'], as_index=False)['order_date'].count()

frequency_df.columns = ['user_id', 'Frequency']
```

```python

frequency_df.head()
```

## Monetary Value
Calculating Monetary Value
Here we are calculating the monetary value of customer spend on purchasing products from the website.

```python

r_cust['Total'] = r_cust['price']*r_cust['quantity']
monetary_df = r_cust.groupby(by='user_id', as_index=False)['Total'].sum()
monetary_df.columns = ['user_id', 'Monetary']
monetary_df.head()
```

Merging all three columns in one dataframe
Here we are merging all the dataframe columns in a single entity using the merge function to display the recency, frequency, monetary value.

```python

rf_df = df_recency.merge(frequency_df, on='user_id')
rfm_df = rf_df.merge(monetary_df, on='user_id')
rfm_df.head()
```

```python
rfm_df.drop(columns ='LastPurchaseDate',inplace=True)
```

```python
rfm_df
```

Ranking Customerâ€™s based upon their recency, frequency, and monetary score
> 1. Recency should be low  min 
> 1. Frequency should be high 
> 1. Monetary should be high

```python

rfm_df['R_rank'] = rfm_df['Recency'].rank(ascending=False)
rfm_df['F_rank'] = rfm_df['Frequency'].rank(ascending=True)
rfm_df['M_rank'] = rfm_df['Monetary'].rank(ascending=True)
```

```python

# normalizing the rank of the customers
rfm_df['R_rank_norm'] = (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm'] = (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm'] = (rfm_df['F_rank']/rfm_df['M_rank'].max())*100
 
rfm_df.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)
 
rfm_df.head()
```
